```python
col = ['Sales_in_thousands', 'year_resale_value',
       'Price_in_thousands', 'Engine_size', 'Horsepower',
       'Wheelbase', 'Fuel_capacity','Fuel_efficiency']
```

```python
for i in col :
  Q1 = df[i].quantile(0.25)
  Q3 = df[i].quantile(0.75)
  IQR = Q3-Q1
  UF = Q3 + 1.5*IQR
  LF = Q1 - 1.5*IQR

  df = df[(df[i]>=LF) & (df[i]<=UF)]
```

## **Step 2 : Model Building**

Single Linear regression ( Single Variable = X variable)

```python
df.columns
```

```python
X = df['Horsepower']  ## Independent variable
Y = df['Fuel_efficiency'] ## Dependent variable
```

```python
# Spliting the data in train and test

from sklearn.model_selection import train_test_split
```

```python
# Spliting the data in 70/30 ratio

x_train,x_test,y_train,y_test = train_test_split (X,Y,test_size = 0.3, random_state = 35)
```

```python
x_train.head(4)
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
## Single Linear regression Model

from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit (x_train.values.reshape(-1,1),y_train.values.reshape(-1,1))
```

```python
y_pred = model.predict (x_test.values.reshape(-1,1))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred,y_test)
```

```python
l = []
for i in range(1000):
  x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3, random_state=i)
  model = LinearRegression()
  model.fit(x_train.values.reshape(-1,1),y_train.values.reshape(-1,1))
  y_pred = model.predict(x_test.values.reshape(-1,1))
  l.append(r2_score(y_pred,y_test))


  print(max(l))
  print(l.index(max(l)))
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3, random_state = 211 )
model = LinearRegression()
model.fit(x_train.values.reshape(-1,1),y_train.values.reshape(-1,1))
y_pred = model.predict(x_test.values.reshape(-1,1))
r2_score(y_pred,y_test)
```
